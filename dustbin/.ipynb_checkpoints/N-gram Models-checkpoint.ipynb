```python
# Reading the Data
with open('swamivivekananda.txt') as f:
    contents = f.read()
    print(contents)
```

## Tokenising The Data

```python
def tokeniser(corpus):
    pattern = "[,.?;]*\w*[\\'-]*\w*"
    list1 = re.findall(pattern,corpus)
    list2 = list(filter("".__ne__,list1))
    return list2
```

```python
tokenised_corpus = tokeniser(contents)
tokenised_corpus = [i.lower() for i in tokenised_corpus]
vocabulary = sorted(list(set(tokenised_corpus)))
print(f"The length of Vocabulary is {len(vocabulary)}")
```

```python
c = ""
for i in tokenised_corpus:
    c = c + " "+ i + " "
preprocessed_corpus = c
```

```python
def countpat(pattern,corpus):
    l = re.findall(pattern,corpus)
    return len(l)
```

```python
for i in range(len(vocabulary)):
    if vocabulary[i] == ".":
        vocabulary[i] = "\."
    elif vocabulary[i] == "?":
        vocabulary[i] = "\?"
    elif vocabulary[i] == ".,":
        vocabulary[i] = "\.,"
```

## Unigram

```python
def prediction():
    prob = np.zeros(len(vocabulary))
    for i in range(len(vocabulary)):
        pat = " "+vocabulary[i]+ " "
        prob[i] = countpat(pat,preprocessed_corpus)/len(tokenised_corpus)
    word = vocabulary[np.argmax(prob)] + " "
    return word
```

```python
def unigram_model(prompt = "",gen_len = 10):
    gen_sen = prompt.strip() + " "
    i = 1
    while i <= gen_len:
        gen_sen += prediction()
        i += 1
    return gen_sen
```

```python
unigram_model(prompt= "The sentence ")
```

### Bigram

```python
def prediction_bi(prev = ""):
    prob = np.zeros(len(vocabulary))
    for i in range(len(vocabulary)):
        prev1 = " " + prev.strip() + " "
        pat = prev1 + " " + vocabulary[i]+ " "
        prob[i] = countpat(pat,preprocessed_corpus)/countpat(prev1, preprocessed_corpus)
    word = vocabulary[np.argmax(prob)]
    return word.strip()
```

```python
def Bigram_model(prompt = "",gen_len = 10):
    prompt = prompt.strip()
    gen_sen = prompt + " "
    sen = prompt.split(" ")
    prev = sen[-1].lower()
    i = 1
    while i <= gen_len:
        pred = prediction_bi(prev) + " "
        gen_sen += pred
        prev = pred
        i += 1
    return gen_sen
```

```python
Bigram_model("The Man have")
```
