### Cleaning data

```python
# Making text corpus without the start and end inverted comma
corpus = ''
for quote in data['quote']:
    corpus += ' ' + quote[1:-1]
corpus = corpus.lower()
```

```python
punctuations = '''!()-[]{};:'"\,<>/?@#$%^&*_~'''
no_punct = ""
for char in corpus:
    if char not in punctuations:
        no_punct = no_punct + char
corpus = no_punct
```

```python
# cleaning text (as all the quotation is started and ended with " then we should remove those)
corpus
```

```python
# Define tokenizer function that is tokenize the given text
def tokenizer(text):
    # Replace numbers from the text
    text = re.sub('\d', '', text)
    # Make tokens
    tokens = text.split(' ')
    return tokens
```

```python
tokens = tokenizer(corpus)
vocabulary = sorted(list(set(tokens)))
# First 10 words in tokens list
print(tokens[0:10])
print(f'\33[34m Total number of tokens presents in corpus is {len(tokens)}')
print(f'\33[31m Length of the vocabulary is {len(vocabulary)}')
```

```python
vocabulary = [x for x in vocabulary if len(x)>1]
vocabulary.remove('th')
```

## Unigram

$$P(w)=\frac{count(w)}{N}$$

```python
def p_unigram(vocabulary):
    counts = []
    for word in vocabulary:
        count = 0
        for token in tokens:
            if word in token:
                count += 1
        counts.append(count)
    return np.array(counts)/len(vocabulary)
```

```python
unigram_probs = p_unigram(vocabulary)
```

```python
def unigram(given_text, vocab, prob):
    new_text = ''
    sorted_idx = np.argsort(prob)
    for i in range(len(sorted_idx)):
        if vocab[sorted_idx[i]].endswith('.'):
            new_text += ' ' + vocab[sorted_idx[i]]
            break
        else:
            new_text += ' ' + vocab[sorted_idx[i]]
    return given_text + new_text
```

```python
unigram('humanity', vocabulary, unigram_probs)
```

```python
def bigram(given_text, pred_len, vocab, corpus):
    for i in tqdm(range(pred_len)):
        last_word = given_text.split()[-1]
        dummy_arr = []
        for word in vocab:
            search_string = last_word + ' ' + word
            dummy_arr.append(corpus.count(search_string)/corpus.count(last_word))
        new_word = vocab[np.argmax(np.array(dummy_arr))]
        given_text += ' ' + new_word
    return given_text
```

### Trigram

$$ P(w_i|w_{i-2}w_{i-1}) = \frac{Count(w_{i-2}w_{i-1}w_i)}{Count(w_{i-2}w_{i-1})}$$

```python
def trigram(given_text, pred_len, vocab, corpus):
    for i in tqdm(range(pred_len)):
        last_word = ''
        for j in given_text.split()[-2:]:
            last_word += ' '+j
        last_word = last_word.strip()
        dummy_arr = []
        for word in vocab:
            search_string = last_word + ' ' + word
            dummy_arr.append(corpus.count(search_string)/corpus.count(last_word))
        new_word = vocab[np.argmax(np.array(dummy_arr))]
        given_text += ' ' + new_word
    return given_text
```

## N-gram

$$ P(w_i|w_{i-n+1}\cdots w_{i-1}) = \frac{Count(w_{i-n+1}\cdots w_{i-1}w_i)}{Count(w_{i-n+1}\cdots w_{i-1})}$$

<div class='alert alert-success'> 
This model multiple limitations, some of them are given below
    
- Due to not exact match of string, count can be zero. Hence division by zero error came up.
- Length of sentence can be fixed.
- Corpus needs close clean up.
- Many more.
    
</div>

```python
def N_gram(given_text, n, pred_len, vocab, corpus):
    for i in tqdm(range(pred_len)):
        last_word = ''
        for j in given_text.split()[-(n-1):]:
            last_word += ' '+j
        last_word = last_word.strip()
        dummy_arr = []
        for word in vocab:
            search_string = last_word + ' ' + word
            dummy_arr.append(corpus.count(search_string)/corpus.count(last_word))
        new_word = vocab[np.argmax(np.array(dummy_arr))]
        given_text += ' ' + new_word
    return given_text
```

```python
N_gram('arise', 2, 5, vocabulary, corpus)
```
